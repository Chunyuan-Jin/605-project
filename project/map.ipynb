```python
sensorRaw = pd.read_csv("p2_test.csv")
sensor=sensorRaw.dropna()
diff = sensor["mean_2018"] - sensor["mean_2017"]
sensor.insert(9, 'diff', diff)
sensor.head()
```

```python
# group by change of pm2.5
pos=sensor[sensor["diff"] >= 0] # 195
neg=sensor[sensor["diff"] < 0]  #156
pos1=pos[pos["diff"] <= pos["diff"].median()]
pos2=pos[pos["diff"] > pos["diff"].median()]
neg1=neg[neg["diff"] >= neg["diff"].median()]
neg2=neg[neg["diff"] < neg["diff"].median()]
```

```python
nodiff=sensor[sensor["p_value"] >= 0.05]
nodiff
```

```python
# define the world map
world_map = folium.Map()

# display world map
world_map

# Sofia latitude and longitude values
latitude = 42.698334
longitude = 23.319941

# Create map and display it
sofia_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# instantiate a feature group for the well_locations in the dataframe
sensor_location = folium.map.FeatureGroup()

# loop through the locationss and add each to the the feature group
for lat, lng in zip(pos1.lat, pos1.lon):
    sensor_location.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='orange',
            fill_color='orange',
            fill_opacity=0.5
        )
    )

# loop through the locationss and add each to the the feature group
for lat, lng in zip(pos2.lat, pos2.lon):
    sensor_location.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='red',
            fill_color='red',
            fill_opacity=0.5
        )
    )

# loop through the locationss and add each to the the feature group
for lat, lng in zip(neg1.lat, neg1.lon):
    sensor_location.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='blue',
            fill_color='blue',
            fill_opacity=0.5
        )
    )

# loop through the locationss and add each to the the feature group
for lat, lng in zip(neg2.lat, neg2.lon):
    sensor_location.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='navy',
            fill_color='navy',
            fill_opacity=0.5
        )
    )

# loop through the locationss and add each to the the feature group
for lat, lng in zip(nodiff.lat, nodiff.lon):
    sensor_location.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='green',
            fill_color='green',
            fill_opacity=0.5
        )
    )


    
sofia_map.add_child(sensor_location)
sofia_map
```

```python
infoRaw = pd.read_csv("sensor_info.csv")
info=infoRaw.dropna()
pressure_diff = info["pressure_2018"] - info["pressure_2017"]
temp_diff = info["temp_2018"] - info["temp_2017"]
hum_diff = info["hum_2018"] - info["hum_2017"]
info.insert(9, 'pressure_diff', pressure_diff)
info.insert(10, 'temp_diff', temp_diff)
info.insert(11, 'hum_diff', hum_diff)
info.head()
```

```python
# group by change of pressure
pressure_pos=info[info["pressure_diff"] >= 0]
pressure_neg=info[info["pressure_diff"] < 0]
pressure_pos1=pressure_pos[pressure_pos["pressure_diff"] <= pressure_pos["pressure_diff"].median()]
pressure_pos2=pressure_pos[pressure_pos["pressure_diff"] > pressure_pos["pressure_diff"].median()]
pressure_neg1=pressure_neg[pressure_neg["pressure_diff"] >= pressure_neg["pressure_diff"].median()]
pressure_neg2=pressure_neg[pressure_neg["pressure_diff"] <= pressure_neg["pressure_diff"].median()]
```

```python
# group by change of temperature
temp_pos=info[info["temp_diff"] >= 0]
temp_neg=info[info["temp_diff"] < 0]
temp_pos1=temp_pos[temp_pos["temp_diff"] <= temp_pos["temp_diff"].median()]
temp_pos2=temp_pos[temp_pos["temp_diff"] > temp_pos["temp_diff"].median()]
temp_neg1=temp_neg[temp_neg["temp_diff"] >= temp_neg["temp_diff"].median()]
temp_neg2=temp_neg[temp_neg["temp_diff"] <= temp_neg["temp_diff"].median()]
```
